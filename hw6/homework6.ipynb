```python
train_data = np.loadtxt('hw2_adaboost_train.txt')
test_data = np.loadtxt('hw2_adaboost_test.txt')

def generate_data(data):
    X = data[:,:-1]
    y = data[:,-1]
    return X,y

X_train, y_train = generate_data(train_data)
X_test, y_test = generate_data(test_data)

def generate_Theta(col, X):
    sorted_X = X[X[:, col].argsort()]
    Theta = []
    for i in range(len(sorted_X)):
        if i == 0:
            Theta.append(sorted_X[i][col] - 1)
        else:
            Theta.append((sorted_X[i][col] + sorted_X[i - 1][col]) / 2)
    return Theta

def GT(theta, s, col):
    def gt(x):
        return 2 * (s * (x[col] - theta) >= 0) - 1
    return gt

def compute_Ein(gt, X, y, Ut):
    Ein = 0
    for i in range(len(X)): 
        Ein += Ut[i] * (gt(X[i]) * y[i] < 0)
    return Ein / len(X)

def update_coef(err):
    return np.sqrt((1 - err) / err)


def get_alpha(coef):
    return math.log(coef, math.exp(1))

def get_best_gt(X_train, y_train, Ut):
    Ein_min = 1000
    theta_best = -1000
    s_best = 1
    gt_best = GT(-1000, 1, 1)
    for col in range(2):
        Theta = generate_Theta(col, X_train)
        for theta in Theta:
            s = 1
            gt = GT(theta, s, col)
            Ein_temp = compute_Ein(gt, X_train, y_train, Ut)
            if Ein_temp < Ein_min or (Ein_temp == Ein_min and np.random.random()>0.5):
                Ein_min = Ein_temp
                theta_best = theta
                gt_best = gt
            s = -1
            gt = GT(theta, s, col)
            Ein_temp = compute_Ein(gt, X_train, y_train, Ut)
            if Ein_temp < Ein_min or (Ein_temp == Ein_min and np.random.random()>0.5):
                Ein_min = Ein_temp
                theta_best = theta
                gt_best = gt
    Ein = compute_Ein(gt_best, X_train, y_train, [1 for i in range(len(X_train))])
    Ut_sum = np.sum(Ut)
    err = np.sum([Ut[i] * (gt_best(X_train[i]) != y_train[i]) for i in range(len(X_train))])/Ut_sum
    next_coef = update_coef(err)
    alpha = get_alpha(next_coef)
    Ut_next = [Ut[i]*((gt_best(X_train[i]) != y_train[i])*next_coef  + (gt_best(X_train[i]) == y_train[i])/next_coef) for i in range(len(X_train))]
    return gt_best, Ein, Ut_next, alpha, Ut_sum, err


```

```python
Ut = [1 / len(X_train) for i in range(len(X_train))]
gt_array = []
alpha_array = []
Ein_array = []
Ut_sum_array = []
err_array = []

for i in range(300):
    gt_best, Ein, Ut, alpha, Ut_sum, err = get_best_gt(X_train, y_train, Ut)
    gt_array.append(gt_best)
    Ein_array.append(Ein)
    alpha_array.append(alpha)
    Ut_sum_array.append(Ut_sum)
    err_array.append(err)
```

```python
def gauss_kernel(gamma, x, x_):
    return math.exp(-gamma * (x - x_).dot(x - x_))

def get_beta(gamma, lam, X, y):
    K = [[gauss_kernel(gamma, xi, xj) for xj in X] for xi in X]
    beta = np.matmul(np.linalg.inv(lam * np.eye(len(X)) + K), y)
    return beta, np.array(K)

def lssvm(beta, K, X, x_n, gamma):
    result = 0
    for i in range(len(X)):
        result += beta[i] * gauss_kernel(gamma, X[i], x_n)
    return result

def compute_Ein_aug(beta, K, lam, y):
    N = len(y)
    E = lam / N * (beta.dot(np.matmul(K, beta))) + \
        1 / N * (np.matmul(K, beta) - y).dot((np.matmul(K, beta) - y))
    return E

                 
def compute_E(beta, K, y):
    E = np.mean(np.sign(np.matmul(K, beta)) != y)
    return E
```
